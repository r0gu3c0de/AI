```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
movie_df = pd.read_csv('Movie_Id_Titles')
movie_df
```

```python
moviesRating_df = pd.read_csv('u.data', sep='\t', names= ['user_id', 'item_id', 'rating', 'timestamp'])
moviesRating_df
```

```python
moviesRating_df.drop(['timestamp'], axis=1, inplace=True)
moviesRating_df
```

```python
moviesRating_df.describe()
```

```python
movies_total_df = pd.merge(moviesRating_df, movie_df, on='item_id')
```

```python
movies_total_df
```

```python
movies_total_df.groupby('title').describe()
```

```python
movies_total_df.groupby('title')['rating'].describe()
```

```python
rating_df_mean = movies_total_df.groupby('title')['rating'].describe()['mean'] 
rating_df_count = movies_total_df.groupby('title')['rating'].describe()['count'] 
```

```python
ratings_df = pd.concat([rating_df_mean, rating_df_count], axis=1)
```

```python
ratings_df
```

```python
ratings_df.reset_index()
```

```python
ratings_df['mean'].plot(bins = 100, kind='hist', color='r')
```

```python
ratings_df['count'].plot(bins = 100, kind='hist', color='g')
```

```python
ratings_df[   ratings_df['mean'] == 5   ]
```

```python
ratings_df.sort_values('count', ascending = False).head(20)
```

```python
userid_movie_matrix = movies_total_df.pivot_table(index = 'user_id', columns= 'title', values= 'rating')
```

```python
userid_movie_matrix
```

```python
star_wars = userid_movie_matrix['Star Wars (1977)']
star_wars
```

```python

star_wars_correlation = pd.DataFrame(userid_movie_matrix.corrwith(star_wars), columns=['Correlation'])
```

```python
star_wars_correlation = star_wars_correlation.join(ratings_df['count'])
```

```python
star_wars_correlation
```

```python
star_wars_correlation.dropna(inplace=True)
star_wars_correlation
```

```python
star_wars_correlation.sort_values('Correlation', ascending=False)
```

```python
star_wars_correlation[star_wars_correlation['count'] >= 80].sort_values('Correlation', ascending=False).head()
```

```python
#Filter for all dataset
userid_movie_matrix
```

```python
movie_correlations = userid_movie_matrix.corr(method = 'pearson', min_periods = 80)
movie_correlations
```

```python
#Based on "preferences"
myRatings = pd.read_csv('My_Ratings.csv')
myRatings
```

```python
similar_movies = pd.Series()

for i in range(0,2):
    similar_movie = movie_correlations[  myRatings['Movie Name'][i]  ].dropna()
    similar_movie = similar_movie.map(lambda x: x*myRatings['Ratings'][i])
    similar_movies = similar_movies.append(similar_movie)
```

```python
similar_movies.sort_values(inplace=True, ascending=False)
```
